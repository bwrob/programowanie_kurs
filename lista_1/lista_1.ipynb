```python
root = bisection(atan_minus_1, 0, 10)
# wolfram alpha: solve arctan(x) - 1 = 0
x = 1.5574077246549022305069748074583602
print(root)
print(abs(x - root))
```

```python
ns = list(range(1, 101))
times = [measure_time(bisection, repeat=1_000)(atan_minus_1, 0.0, 10**n) for n in ns]

plt.plot(ns, times)
plt.xlabel("n")
plt.ylabel("Czas wykonania (ns)")
plt.title("Czas wykonania metody bisekcji dla Pn")
plt.grid(True)
plt.show()
```
